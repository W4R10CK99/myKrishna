```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader,ServiceContext,PromptTemplate
```

```python
document = SimpleDirectoryReader("Source").load_data()
```

```python
document
```

```python
# converting the document into llama index
index = VectorStoreIndex.from_documents(document, show_progress=True)
```

```python
index
```

```python
query_engine = index.as_query_engine()
```

```python
# it is now a retriever, whatever is asked will be retrieved
query_engine
```

```python
response = query_engine.query("What is Karma")
```

```python
print(response)
```

```python
response2 = query_engine.query("How to be happy in life")
print(response2)
```

```python
import os.path
from llama_index.core import (
    VectorStoreIndex,
    SimpleDirectoryReader,
    StorageContext,
    load_index_from_storage,
)

#check if the knowledgeBase already exists
KB_DIR = "./knowledge_base"
if not os.path.exists(KB_DIR):
    #knowledgebase is not present, hence perform indexing and store knowledge
    documents = SimpleDirectoryReader("Source").load_data()
    index = VectorStoreIndex.from_documents(documents)
    
    #store it for later use
    index.storage_context.persist(persist_dir = KB_DIR)
    
else:
    storage_context = StorageContext.from_defaults(persist_dir = KB_DIR)
    index = load_index_from_storage(storage_context)
```
